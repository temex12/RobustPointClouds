```python
import pandas as pd

# Define the path to the CSV files
path_20e_05 = 'l2_norms_final.csv'
path_10e_05 = 'l2_norms_final_10e.csv'
path_20e_02 = 'l2_norms_final_20e_02.csv'

# Load the data
data_20e_05 = pd.read_csv(path_20e_05)
data_10e_05 = pd.read_csv(path_10e_05)
data_20e_02 = pd.read_csv(path_20e_02)

# Assuming 'L2 Norm' is the correct column name, we will proceed to extract this column and calculate descriptive statistics
if 'L2 Norm' in data_20e_05.columns:
    summary_20e_05 = data_20e_05['L2 Norm'].describe()
else:
    summary_20e_05 = 'Column not found in 20 epochs, 0.05 data'

if 'L2 Norm' in data_10e_05.columns:
    summary_10e_05 = data_10e_05['L2 Norm'].describe()
else:
    summary_10e_05 = 'Column not found in 10 epochs, 0.05 data'

if 'L2 Norm' in data_20e_02.columns:
    summary_20e_02 = data_20e_02['L2 Norm'].describe()
else:
    summary_20e_02 = 'Column not found in 20 epochs, 0.02 data'

summary_20e_05, summary_10e_05, summary_20e_02
```

```python
import pandas as pd

# Load your CSV data into a DataFrame
df = pd.read_csv('l2_norms_final.csv')

# Calculate basic statistics
mean_l2 = df['L2 Norm'].mean()
median_l2 = df['L2 Norm'].median()
std_l2 = df['L2 Norm'].std()
min_l2 = df['L2 Norm'].min()
max_l2 = df['L2 Norm'].max()

print(f"Mean L2 Norm: {mean_l2}")
print(f"Median L2 Norm: {median_l2}")
print(f"Standard Deviation: {std_l2}")
print(f"Min L2 Norm: {min_l2}")
print(f"Max L2 Norm: {max_l2}")
```

```python
import matplotlib.pyplot as plt

# Histogram of L2 norms
plt.figure(figsize=(10, 6))
plt.hist(df['L2 Norm'], bins=50, color='blue', alpha=0.7)
plt.title('Histogram of L2 Norms')
plt.xlabel('L2 Norm')
plt.ylabel('Frequency')
plt.show()

# Line plot for L2 norms across batches/epochs if sequential data is available
plt.figure(figsize=(10, 6))
plt.plot(df['L2 Norm'], color='red')
plt.title('L2 Norms Over Batches')
plt.xlabel('Batch Index')
plt.ylabel('L2 Norm')
plt.show()
```
